## <font color='orange'>Removing Outliers</font>

```python
lower_q = df["UN_Estimate"].quantile(0.25)
lower_q
```

```python
higher_q = df["UN_Estimate"].quantile(0.75)
higher_q
```

```python
# inner quantile = upper q- lower q
iqr = higher_q - lower_q
```

```python
print(iqr)
```

```python
lower_q - 1.5 *iqr
# 2039 -1.5 * 18700
```

```python
higher_q + 1.5 * iqr
# 20740 + 1.5 * 18700
```

```python
df_filtered = df[(df["UN_Estimate"] > lower_q - 1.5 * iqr ) & (df["UN_Estimate"] < higher_q + 1.5* iqr )]
```
